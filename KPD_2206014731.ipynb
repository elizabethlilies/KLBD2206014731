1) Number Guess Games

The computer will come up with a random number between 1 and 1000. The player must then continue to guess numbers until the player guesses the correct number. For every guess, the computer will either say "Too high" or "Too low", and then ask for another input. At the end of the game, the number is revealed along with the number of guesses it took to get the correct number.

```python
i = int(1)
print('Guess a number : ')
n = int(input())
if (n==781):
  print('The correct number = 781')
  print('The number of guess = 1')
else:
    while n!=781:
      if (n<781):
        print('too low')
      else:
        print('too high')
        print('Guess a number (again)')
        n = int(input())
        i = i + 1
    print('The correct number = 781')
    print('The number of guess = ', i)
```

1) An algorithm to compute the mean and variance of 1000 data

```python
from random import random
a = int(0)
c = int(0)
for i in range(1,1001):
  b=100*random()
  while b<70:
    b=100*random()
  print('nilai ke', i, '=', b)
  c=c+b**2
  a=a+b
rat=a/1000
var=c/1000-(a/1000)**2
print('Nilai rata-rata = ', rat)
print('Nilai variansi = ', var)
```
